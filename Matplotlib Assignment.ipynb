## What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is used for data visualization and analysis purposes, and it is often used to visualize scientific, engineering, and mathematical data in a graphical format. Matplotlib is also used in data science, machine learning, and artificial intelligence applications, among others.

The Pyplot module of Matplotlib provides a simple interface to create a variety of plots. Here are five plots that can be created using Pyplot:

1.Line Plot: A line plot is a graph that displays data as a series of points connected by straight lines. It is used to show trends over time or to compare two or more sets of data.

2.Scatter Plot: A scatter plot is a graph that displays the relationship between two variables by plotting points on a two-dimensional plane. It is used to identify patterns and relationships in the data.

3.Bar Plot: A bar plot is a graph that displays data as rectangular bars. It is used to compare data across different categories or groups.

4.Histogram: A histogram is a graph that displays the distribution of a set of data. It is used to visualize the shape of the data and to identify any outliers or anomalies.

5.Pie Chart: A pie chart is a graph that displays data as a circle divided into segments. It is used to show the relative proportions of different categories or groups.

## What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

A scatter plot is a graph that shows the relationship between two numerical variables in a dataset. It displays data points as individual dots or markers on a two-dimensional plot, with each dot representing a single observation in the dataset. Scatter plots are used to visually identify any patterns or relationships between the variables and are commonly used in exploratory data analysis and statistical modeling.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.scatter(x,y,c='#4708FC')
plt.xlabel('This is my x axis')
plt.ylabel("This is my y axis")
plt.title("Scatter Plot")
```

# Q 3.

## Why is the subplot() function used? Draw four line plots using the subplot() function.
### Use the following data:

```python
fig,axs = plt.subplots(nrows=2, ncols=2)
axs[0,0].plot( np.array([0, 1, 2, 3, 4, 5]), np.array([0, 100, 200, 300, 400, 500]))
axs[0,0].set_title("Line 1")
axs[0,1].plot( np.array([0, 1, 2, 3, 4, 5]) , np.array([50, 20, 40, 20, 60, 70]))
axs[0,1].set_title("Line 2")
axs[1,0].plot( np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60]))
axs[1,0].set_title("Line 3")
axs[1,1].plot( np.array([0, 1, 2, 3, 4, 5]),  np.array([200, 350, 250, 550, 450, 150]))
axs[1,1].set_title("Line 4")
plt.show()
```

## What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is a type of plot used to display the distribution or comparison of categorical data. It consists of a series of bars, where the height of each bar represents the frequency, count, or percentage of observations in each category.

Bar plots are commonly used to:

-->Compare the frequency or count of categorical data across different groups or categories.

-->Display the proportion or percentage of observations in each category.

-->Show the change in the frequency or count of categorical data over time or across different conditions.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
```

```python
#Horizontal bar plot
plt.barh(company,profit)
```

# Q 5.

## What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

fig, axs = plt.subplots(ncols=2,)
axs[0].boxplot(box1)
axs[0].set_title('Box 1')
axs[1].boxplot(box2)
axs[1].set_title('Box 2')
plt.show()
```
